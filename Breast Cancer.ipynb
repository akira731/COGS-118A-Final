```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

def run_trials(X, y, partitions, n_trials=3):
    # Classifiers to test
    classifiers = {
        'Random Forest': RandomForestClassifier(random_state=42),
        'KNN': KNeighborsClassifier(n_neighbors=5),
        'SVM': SVC(kernel='rbf', random_state=42)
    }
    
    # Store results
    results = {
        'Random Forest': {'train_accuracies': [], 'test_accuracies': []},
        'KNN': {'train_accuracies': [], 'test_accuracies': []},
        'SVM': {'train_accuracies': [], 'test_accuracies': []}
    }
    
    # Run trials for each partition
    for partition in partitions:
        print(f"\nPartition: {partition}")
        for trial in range(n_trials):
            # Set a different random seed for each trial
            random_seed = 42 + trial
            
            # Split the data based on the current partition
            if partition == '80/20':
                X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_seed)
            elif partition == '50/50':
                X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=random_seed)
            else:  # '20/80'
                X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=random_seed)
            
            # Test each classifier
            for name, classifier in classifiers.items():
                # Train the classifier
                clf = classifier
                clf.fit(X_train, y_train)
                
                # Compute accuracies
                train_acc = clf.score(X_train, y_train)
                test_acc = clf.score(X_test, y_test)
                
                # Store results
                results[name]['train_accuracies'].append(train_acc)
                results[name]['test_accuracies'].append(test_acc)
                
                # Print individual trial results
                print(f"{name} - Trial {trial+1} - Train Accuracy: {train_acc:.4f}, Test Accuracy: {test_acc:.4f}")
    
    # Compute and print summary statistics
    print("\nSummary Statistics:")
    for name, result in results.items():
        print(f"\n{name}:")
        print("Train Accuracies:")
        print(f"  Mean: {np.mean(result['train_accuracies']):.4f}")
        print(f"  Std Dev: {np.std(result['train_accuracies']):.4f}")
        print("Test Accuracies:")
        print(f"  Mean: {np.mean(result['test_accuracies']):.4f}")
        print(f"  Std Dev: {np.std(result['test_accuracies']):.4f}")
    
    return results

partitions = ['20/80', '50/50', '80/20']
results = run_trials(X, y, partitions)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler

def hyperparameter_tuning(X, y):
    # Define partitions
    partitions = [
        {'train_size': 0.8, 'test_size': 0.2},
        {'train_size': 0.5, 'test_size': 0.5},
        {'train_size': 0.2, 'test_size': 0.8}
    ]
    
    # Hyperparameter grids for each classifier
    param_grids = {
        'Random Forest': {
            'n_estimators': [50, 100, 200],
            'max_depth': [None, 10, 20, 30],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 2, 4]
        },
        'KNN': {
            'n_neighbors': [3, 5, 7, 9],
            'weights': ['uniform', 'distance'],
            'algorithm': ['auto', 'ball_tree', 'kd_tree']
        },
        'SVM': {
            'C': [0.1, 1, 10, 100],
            'kernel': ['linear', 'rbf'],
            'gamma': ['scale', 'auto', 0.1, 1]
        }
    }
    
    # Results storage
    results = {}
    
    # Standardize features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # Iterate through partitions
    for partition_idx, partition in enumerate(partitions, 1):
        print(f"\n--- Partition {partition_idx}: {partition['train_size']*100}/{partition['test_size']*100} Split ---")
        
        # Store results for this partition
        results[f'Partition_{partition_idx}'] = {}
        
        # Perform 3 trials for each partition
        for trial in range(3):
            # Set random seed for reproducibility
            np.random.seed(42 + trial)
            
            # Split the data
            X_train, X_test, y_train, y_test = train_test_split(
                X_scaled, y, 
                train_size=partition['train_size'], 
                test_size=partition['test_size'], 
                random_state=42 + trial
            )
            
            # Iterate through classifiers
            for clf_name, clf_class in [
                ('Random Forest', RandomForestClassifier()),
                ('KNN', KNeighborsClassifier()),
                ('SVM', SVC())
            ]:
                # Perform Grid Search
                grid_search = GridSearchCV(
                    estimator=clf_class,
                    param_grid=param_grids[clf_name],
                    cv=5,  # 5-fold cross-validation
                    scoring='accuracy',
                    n_jobs=-1  # Use all available cores
                )
                
                # Fit grid search
                grid_search.fit(X_train, y_train)
                
                # Best model with tuned hyperparameters
                best_model = grid_search.best_estimator_
                
                # Compute accuracies
                train_acc = best_model.score(X_train, y_train)
                test_acc = best_model.score(X_test, y_test)
                
                # Store results
                result_key = f'{clf_name}_Trial_{trial+1}'
                results[f'Partition_{partition_idx}'][result_key] = {
                    'Best Hyperparameters': grid_search.best_params_,
                    'Best Cross-Validation Score': grid_search.best_score_,
                    'Train Accuracy': train_acc,
                    'Test Accuracy': test_acc
                }
                
                # Print results
                print(f"\n{clf_name} - Trial {trial+1}")
                print(f"Best Hyperparameters: {grid_search.best_params_}")
                print(f"Best CV Score: {grid_search.best_score_:.4f}")
                print(f"Train Accuracy: {train_acc:.4f}")
                print(f"Test Accuracy: {test_acc:.4f}")
    
    return results
    
results = hyperparameter_tuning(X, y)

def results_to_dataframe(results):
    data = []
    for partition, classifiers in results.items():
        for clf_name, details in classifiers.items():
            data.append({
                'Partition': partition,
                'Classifier': clf_name,
                'Best Hyperparameters': str(details['Best Hyperparameters']),
                'Best CV Score': details['Best Cross-Validation Score'],
                'Train Accuracy': details['Train Accuracy'],
                'Test Accuracy': details['Test Accuracy']
            })
    return pd.DataFrame(data)

# Convert results to DataFrame
results_df = results_to_dataframe(results)
print("\nResults Summary:")
print(results_df)
```
